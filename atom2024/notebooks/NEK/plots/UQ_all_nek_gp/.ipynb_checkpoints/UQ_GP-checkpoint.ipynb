# Swarm Plots for GP models 

```python

# Directory path
dir_path = '/Users/radhi/Desktop/CAPSTONE_DATA/GP_result_df'

# List all files in the directory
all_files = os.listdir(dir_path)

# Filter files containing "test" in their names
test_files = [f for f in all_files if 'test' in f.lower() and f.endswith('.csv')]

# Initialize a dictionary to store DataFrames
dataframes = {}

# Loop through the filtered files and read each into a DataFrame
for file in test_files:
    file_path = os.path.join(dir_path, file)
    df = pd.read_csv(file_path)
    df_name = os.path.splitext(file)[0]  # Get the file name without the extension
    dataframes[df_name] = df
    print(f"Read {file} into a DataFrame with name '{df_name}'")
  
```

```python
df = dataframes['NEK5_binding_moe_SMOTE_test_GP']

```

```python
# Load your data
data = df

pred_prob_class0 = torch.tensor(data['pred_prob_class0'].values)
pred_prob_class1 = torch.tensor(data['pred_prob_class1'].values)

# Stack the probabilities to create a distribution
prob_dist = torch.stack([pred_prob_class0, pred_prob_class1], dim=-1)

# Draw samples from the posterior distribution
pred_samples = prob_dist.unsqueeze(0).repeat(256, 1, 1).exp()

# Normalize the samples
probabilities = (pred_samples / pred_samples.sum(-2, keepdim=True)).mean(0)

# Add the probabilities back to the DataFrame
data['normalized_prob_class0'] = probabilities[:, 0].numpy()
data['normalized_prob_class1'] = probabilities[:, 1].numpy()

# Create the swarm plot using normalized probabilities for class 1
plt.figure(figsize=(10, 6))
sns.swarmplot(x='prediction_type', y='normalized_prob_class1', data=data)

# Customize the plot
plt.title('Swarm Plot of Normalized Predicted Probabilities by Prediction Type')
plt.xlabel('Prediction Type')
plt.ylabel('Normalized Predicted Probability')
plt.show()
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch

# Directory path
dir_path = '/Users/radhi/Desktop/CAPSTONE_DATA/GP_result_df'

# List all files in the directory
all_files = os.listdir(dir_path)

# Filter files containing "test" in their names
test_files = [f for f in all_files if 'test' in f.lower() and f.endswith('.csv')]

# Initialize a dictionary to store DataFrames
dataframes = {}

# Loop through the filtered files and read each into a DataFrame
for file in test_files:
    file_path = os.path.join(dir_path, file)
    df = pd.read_csv(file_path)
    df_name = os.path.splitext(file)[0]  # Get the file name without the extension
    dataframes[df_name] = df
    print(f"Read {file} into a DataFrame with name '{df_name}'")

# Loop through each DataFrame to create individual swarm plots
for df_name, df in dataframes.items():
    # Check if the necessary columns are present
    if 'pred_prob_class0' in df.columns and 'pred_prob_class1' in df.columns:
        # Convert the columns to tensors
        pred_prob_class0 = torch.tensor(df['pred_prob_class0'].values)
        pred_prob_class1 = torch.tensor(df['pred_prob_class1'].values)
        
        # Stack the probabilities to create a distribution
        prob_dist = torch.stack([pred_prob_class0, pred_prob_class1], dim=-1)
        
        # Draw samples from the posterior distribution
        pred_samples = prob_dist.unsqueeze(0).repeat(256, 1, 1).exp()
        
        # Normalize the samples
        probabilities = (pred_samples / pred_samples.sum(-2, keepdim=True)).mean(0)
        
        # Add the probabilities back to the DataFrame
        df['normalized_prob_class0'] = probabilities[:, 0].numpy()
        df['normalized_prob_class1'] = probabilities[:, 1].numpy()
        
        # Create the swarm plot using normalized probabilities for class 1
        plt.figure(figsize=(12, 8))
        sns.swarmplot(x='prediction_type', y='normalized_prob_class1', data=df)

        # Customize the plot
        plt.title(f'Swarm Plot of Normalized Predicted Probabilities by Prediction Type for {df_name}')
        plt.xlabel('Prediction Type')
        plt.ylabel('Normalized Predicted Probability')
        plt.tight_layout()
    
        plt.show()
    else:
        print(f"Skipping {df_name} because it does not contain the necessary columns")
```
